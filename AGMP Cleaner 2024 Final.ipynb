```python
#import pandas library
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
#read in csv file from directory
agmp_data = pd.read_csv("/Users/perceval/data_science/agmp/data_AGMP_19July2022_v2.csv")
```

```python
#display contents of the csv file

##### renaming columns ############

# Pubmed_id = Publication_ID
# phenotype = phenT
agmp_data
```

```python
# convert the csv file into a data frame
df = pd.DataFrame(agmp_data)
```

```python
# display the contents of the dataframe
df
```

```python

# split the column "origin_of_participants" into columns separated by a ,
split_values = df['origin_of_participants'].str.split(',', expand=True)
```

```python
# display the split values
split_values
```

```python

# rename the column names
split_values.columns = [f'country_participant_0{i+1}' for i in range(split_values.shape[1])]
```

```python
split_values.columns
```

```python

# Concatenate the split with the original data frame
result_df = pd.concat([df, split_values], axis=1)
```

```python
result_df.head()
```

```python
result_df.columns
```

```python

result_df['latitude_01'] = ''
result_df['longitude_01'] = ''
result_df['latitude_02'] = ''
result_df['longitude_02'] = ''
result_df['latitude_03'] = ''
result_df['longitude_03'] = ''
result_df['latitude_04'] = ''
result_df['longitude_04'] = ''
result_df['latitude_05'] = ''
result_df['longitude_05'] = ''
result_df['latitude_06'] = ''
result_df['longitude_06'] = ''
result_df['latitude_07'] = ''
result_df['longitude_07'] = ''
```

```python
result_df.columns
```

```python
df = result_df.iloc[:200000] 
```

```python
df.shape
df
```

```python
geolocator = Nominatim(user_agent="myApp01")

for i in df.index:
    try:
        #tries fetch address from geopy
        location = geolocator.geocode(df['country_participant_01'][i])
        
        #append lat/long to column using dataframe location
        df.loc[i,'latitude_01'] = location.latitude
        df.loc[i,'longitude_01'] = location.longitude

    except:
        #catches exception for the case where no value is returned
        #appends null value to column
        df.loc[i,'latitude_01'] = ""
        df.loc[i,'longitude_01'] = ""
      

 
```

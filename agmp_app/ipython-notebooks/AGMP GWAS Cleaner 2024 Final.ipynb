```python
#improve to make results retrieved faster
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import pandas as pd

# Initialize the geolocator with a user agent and rate limiter
geolocator = Nominatim(user_agent="myApp01")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)  # adjust delay to respect usage policy

# Assuming 'country_participant_01' contains the addresses
addresses = agmp_data_wip['country_participant_01'].unique()  # Process each unique address once

# Use a dictionary to cache geocoding results
location_cache = {}

# Pre-fetch locations for unique addresses
for address in addresses:
    if address not in location_cache:  # Check if address is already geocoded
        location = geocode(address)
        if location:
            location_cache[address] = (location.latitude, location.longitude)
        else:
            location_cache[address] = (None, None)

# Map the results back to the DataFrame without using a loop
agmp_data_wip['latitude_01'] = agmp_data_wip['country_participant_01'].map(lambda x: location_cache[x][0])
agmp_data_wip['longitude_01'] = agmp_data_wip['country_participant_01'].map(lambda x: location_cache[x][1])

# Handle missing data
agmp_data_wip[['latitude_01', 'longitude_01']] = agmp_data_wip[['latitude_01', 'longitude_01']].fillna("")
```
